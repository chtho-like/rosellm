```python
import torch
import torch.nn as nn
import torch.nn.functional as F

# Define the input tensor.
batch_size = 2
seq_len = 3
hidden_size = 4
x = torch.randn(batch_size, seq_len, hidden_size)
print("Input tensor shape:", x.shape)
print("Input tensor:\n", x)

# PyTorch's built-in LayerNorm implementation.
layer_norm_pytorch = nn.LayerNorm(hidden_size)
output_pytorch = layer_norm_pytorch(x)
print("\nPyTorch LayerNorm output:\n", output_pytorch)

# Manual implementation of LayerNorm.
def manual_layer_norm(x, eps=1e-5):
    """
    Manual implementation of LayerNorm without using PyTorch's built-in functions.
    
    Args:
        x: Input tensor of shape [batch_size, seq_len, hidden_size].
        eps: Small constant for numerical stability.
    
    Returns:
        Normalized tensor of the same shape as input.
    """
    # Calculate mean along the last dimension.
    mean = x.mean(dim=-1, keepdim=True)
    
    # Calculate variance along the last dimension.
    var = ((x - mean) ** 2).mean(dim=-1, keepdim=True)
    
    # Normalize the input.
    x_norm = (x - mean) / torch.sqrt(var + eps)
    
    # In a complete implementation, we would also have learnable parameters
    # gamma (scale) and beta (shift), but for simplicity, we're omitting them.
    # If needed, they would be applied as: x_norm * gamma + beta
    
    return x_norm

# Apply our manual LayerNorm.
output_manual = manual_layer_norm(x)
print("\nManual LayerNorm output:\n", output_manual)

# Complete manual implementation with learnable parameters.
class ManualLayerNorm(nn.Module):
    """
    Complete manual implementation of LayerNorm with learnable parameters.
    """
    def __init__(self, normalized_shape, eps=1e-5):
        """
        Initialize the LayerNorm module.
        
        Args:
            normalized_shape: Size of the last dimension of input tensor.
            eps: Small constant for numerical stability.
        """
        super(ManualLayerNorm, self).__init__()
        self.eps = eps
        self.gamma = nn.Parameter(torch.ones(normalized_shape))
        self.beta = nn.Parameter(torch.zeros(normalized_shape))
    
    def forward(self, x):
        """
        Forward pass of the LayerNorm module.
        
        Args:
            x: Input tensor of shape [..., normalized_shape].
        
        Returns:
            Normalized tensor of the same shape as input.
        """
        mean = x.mean(dim=-1, keepdim=True)
        var = ((x - mean) ** 2).mean(dim=-1, keepdim=True)
        x_norm = (x - mean) / torch.sqrt(var + self.eps)
        
        # Apply scale and shift.
        return x_norm * self.gamma + self.beta

# Create and apply our complete manual LayerNorm.
manual_ln_module = ManualLayerNorm(hidden_size)
output_manual_module = manual_ln_module(x)
print("\nManual LayerNorm with learnable parameters output:\n", output_manual_module)

# Verify that our implementation matches PyTorch's.
print("\nDifference between PyTorch and our manual implementation:")
print(torch.abs(output_pytorch - output_manual_module).max().item())
```
